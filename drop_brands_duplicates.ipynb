```python
import pandas as pd
import numpy as np
```

```python
def stand(a):
    a = str(a)
    a = a.replace(' ', '')
    a = a.replace('.', '')
    a = a.replace(',', '')
    a = a.lower()
    return a
```

```python
data = pd.read_csv('data/cases.csv')
data['Тест'] = data['Компания'].apply(stand)
data = data.drop_duplicates()
data['Длинна'] = data['Тест'].apply(len)
data = data.sort_values('Длинна', axis=0, ascending=True, ignore_index=True)
data
```
