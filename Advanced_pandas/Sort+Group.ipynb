```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv', sep=',')
```

```python
melb_df.head()
melb_df.info()
```

```python
melb_df['Date'] = melb_df['Date'].astype('datetime64')
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
melb_df[melb_df['Date'].dt.quarter == 2].count()
```

```python
exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in exclude:
        melb_df[col] = melb_df[col].astype('category')
melb_df.info()
```

```python
melb_df.sort_values(by='Price').head(10)
```

```python
melb_df.sort_values(by = 'Date', ascending = False)
```

```python
# Отсортируем таблицу сначала по возрастанию расстояния от центра города (Distance), а затем — по возрастанию цены объекта (Price). 
# Для того чтобы вывод был более наглядным, выделим каждую десятую строку из столбцов Distance и Price результирующей таблицы

melb_df.sort_values(by = ['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
# Хронологию продаж таунхаусов, у которых площадь участка существенно больше площади здания, 
# чтобы понять, как часто компания справляется с таким сложным видом объектов. 
# Объекты, проданные в один и тот же день, мы бы хотели сортировать по значению коэффициента соотношения площадей. 

mask_1 = melb_df['AreaRatio'] < -0.8
mask_2 = melb_df['Type'] == 'townhouse'
mask_3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask_1 & mask_2 & mask_3].sort_values(by = ['Date', 'AreaRatio'],
                                              ascending = [True, False],
                                              ignore_index=True).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.groupby(by = 'Type', as_index = False).mean()
```

```python
melb_df.groupby('Type')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
```

```python
melb_df.groupby('MonthSale')['Price'].agg(['count', 'mean', 'max']).sort_values(by = 'count', ascending = False)
```
